```python
def factorial(n):
    if n==1 or n==0:
        return 1
    else: 
        return (n* factorial(n-1))

def perm_no_repeat(n,r):
    return factorial(n)//factorial(n-r)

def perm_repeat(n,r):
    return n**r

def comb_no_repeat(n,r):
    return factorial(n)//(factorial(r)*factorial(n-r))

def comb_repeat(n,r):
    return factorial(n+r-1)//(factorial(r)*factorial(n-1))
```

```python
# Определяем случайные значения параметров n и k
n = np.random.randint(1, 8)
k = np.random.randint(1, 8)

# Определяем число M случайных реализаций набора n1, ..., nk
M = 10**4

# Определяем искомый набор биномиальных коэффициентов, полученных в цифровых экспериментах
C = np.zeros(M)

# Функция для вычисления биномиального коэффициента
def binomial_coefficient(n, n_k):
    b = 1
    for i in n_k:
        b=b*factorial(i)
    return factorial(n) / b

# Запускаем цикл случайных реализаций набора n1, ..., nk
for m in range(M):
    # Подбираем набор n_k = (n1, ..., nk), который удовлетворяет условию n1 + ... + nk = n
    while True:
        n_k = np.random.rand(k)
        n_k = n * n_k / np.sum(n_k)
        for i in range(k):
            n_k[i] = np.floor(n_k[i]) if np.random.randint(2) == 0 else np.round(n_k[i])
        
        if np.sum(n_k) == n:
            break
    
    # Вычисляем соответствующий биномиальный коэффициент
    C[m] = binomial_coefficient(n, n_k)

# Рисуем гистограмму частот появления биномиальных коэффициентов
plt.hist(C, bins=np.arange(min(C), max(C)+1), edgecolor='black')
plt.xlabel(f'Биномиальные коэффициенты n ={n}, k = {k}')
plt.ylabel('Частота')
plt.title('Частота появления биномиальных коэффициентов')
plt.show()
```

```python
# Определяем число благ, максимальную цену одного из благ,
# цену комплекта из трех благ
N = 1000
P_max = 10000
P = 1500

# Определяем случайным образом набор цен всех благ,
# ранжируем цены по возрастанию
Pr = np.random.randint(1, P_max+1, N)
Pr.sort()

# Запускаем цикл перебора всех комплектов из трех благ
res = []
k = 0

# Найдем индекс, начиная с которого сумма будет точно больше
max_i = -1
for i in range(N):
    if (Pr[i]>P):
        max_i = i-1
        break

# Идем слева
for i1 in range(max_i):
    # Идем справа
    for i2 in range(max_i, i1+1, -1):
        
        if (Pr[i1] + Pr[i2] < P):
            # Поиск бинарный, так как упорядоченный набор
            l,r = i1,i2
            diff = P - (Pr[i1] + Pr[i2])
            while (r-l>1):
                mid = (l+r)//2
                if (Pr[mid]<= diff):
                    l = mid
                else:
                    r = mid
            i3 = l
            if Pr[i1] + Pr[i2] + Pr[i3] == P:
                k += 1
                # Выводим номер комплекта из трех благ, а также
                # номера благ, сумма цен которых равняется
                # известной величине
                res.append([i1,i2,i3,Pr[i1],Pr[i2],Pr[i3]])

res = np.array(res)

df = pd.DataFrame({'i1': res[::,0],'p1': res[::,3], 'i2': res[::,1],'p2': res[::,4], 'i3': res[::,2],'p3': res[::,5], 'sum': res[::,3]+ res[::,4] + res[::,5]})
df
```

```python
def sub_factorial(k):
    m = [1,-1]
    k_sum = factorial(k)
    for i in range(1,k+1):
        k_sum += m[i%2]*factorial(k)//factorial(i)
        
    return k_sum

n,k = 5,4
comb_no_repeat(n+k,n) * sub_factorial(k)
```
