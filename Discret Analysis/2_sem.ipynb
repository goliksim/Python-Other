```python
# Определим функцию f(x)
def fnm(x, n,m):
    return (n+m)**((m+n)/n) * (1-x**n)*(x**m) / (n*m**(m/n))

x = np.linspace(0, 1, 100)


# Построим график функции
for i in range(25):
    n, m = 3* random(), 3* random()
    plt.plot(x, fnm(x,n,m))

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(' Графики функции f_nm для 25 пар случайных значений параметров n и m')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```

**q<sub>1</sub> = a<sub>1</sub> + √2b<sub>1</sub> + √3c<sub>1</sub> + √6d<sub>1</sub>** и

Таблица проверки действий гомоморфизма, обеспечивающего отображение f: ℤ<sub>4</sub> → ℤ'<sub>4</sub>, содержит 16 возможных пар значений (n,m).

```python
def check_variant(variant, n, m):
    f1 = variant[(n+m) % 4]
    f2 = (3*variant[n] + 5*variant[m]) % 4
    return f1 == f2


variants = [0, 1, 2, 3]
result = []

for variant in itertools.product(variants, variants, variants, variants):
    check = True
    for (n, m) in itertools.product(variants, variants):
        if not check_variant(variant, n, m):
            check = False
            break
    if check:
        result.append(variant)

def print_turple(array):
    return [" ".join([str(s) for s in x]) for x in array] 

print("Найденные гомоморфизмы:\n\n","\n\n".join(print_turple(result)), sep='')
    
```

Cвойство ассоциативности операции (23):</br>
**i ∗ (j ∗ k) = (i ∗ j) ∗ k**

```python
def k_multiply(i,j):
    return 1+(i+j-2)%33


def check_concurrence(numbers):
    i,j,k = numbers
    f1 = k_multiply(i,k_multiply(j,k))
    f2 = k_multiply(k_multiply(i,j),k)
    return f1==f2

counter = 0
for number_set in itertools.product(range(1,34),range(1,34),range(1,34)):
    if (check_concurrence(number_set)):
        counter +=1

print("Количество совпадений: ",counter)

print("Алгебраическая операция обладает свойством ассоциативности: ",counter == 33**3)
```

```python
def k_multiply_2(i,j):
    return 1+(i-1)*(j-1)%33


def check_concurrence_2(numbers):
    i,j,k = numbers
    f1 = k_multiply_2(i,k_multiply_2(j,k))
    f2 = k_multiply_2(k_multiply_2(i,j),k)
    return f1==f2

counter = 0
for number_set in itertools.product(range(1,34),range(1,34),range(1,34)):
    if (check_concurrence_2(number_set)):
        counter +=1

print("Количество совпадений: ",counter)
print("Алгебраическая операция обладает свойством ассоциативности: ",counter == 33**3)
```

Проверить свойство ассоциативности по умножению и дистрибутивности.

```python
def check_20(x,y):
    return np.array([x[0] * y[0] + 2*x[1]*y[1], x[1]*y[0] + x[0]*y[1]])

x, y, z  = [random.randint(1, 100) for _ in range(2)], [random.randint(1, 100) for _ in range(2)], [random.randint(1, 100) for _ in range(2)]
x, y, z = np.array(x), np.array(y), np.array(z)

#Проверяем свойство ассоциативности путем вычитания
#правой части из левой соответствующих уравнений
left = check_20(check_20(x,y),z)
right = check_20(x, check_20(y,z))
ass = left - right

#Проверяем свойство дистрибутивности слева и справа по
#сложению аналогично предыдущему случаю
left = check_20(x+y,z)
right = check_20(x,z) + (check_20(y,z))
distr_L = left - right

left = check_20(x,y+z)
right = check_20(x,y) + (check_20(x,z))
distr_R = left - right

print('ass = ', ass)
print('distr_L = ', distr_L)
print('distr_R = ', distr_R)
```

Нули означают, что после вычитания</br>
правой части уравнений из левой части соответствующих уравнений получен</br>
ноль, как для случая проверки свойства ассоциативности операции умножения, так и свойства дистрибутивности слева и справа по сложению. </br>

```python
# Определяем максимально возможную размерность
n_max = 500
# Создаем массив для хранения норм
norm_s = np.zeros(n_max)

for n in range(1, n_max + 1):
    # Генерируем n векторов с n координатами, равномерно
    # случайно распределенными на отрезке [-1, 1]
    x = -1 + 2 * np.random.rand(n, n)
    # Нормируем длину каждого вектора на единицу
    for i in range(n):
        x[i, :] = x[i, :] / np.linalg.norm(x[i, :])
    # Подсчитываем сумму векторов s
    s = np.sum(x, axis=0)
    # Находим длину вектора s
    norm_s[n - 1] = np.linalg.norm(s)

# Рисуем искомую зависимость длины вектора s от размерности
# пространства n
print('График зависимости длины вектора суммы |s| от размерности пространства n')
plt.plot(range(1, n_max + 1), norm_s, '*')
plt.xlabel('n')
plt.ylabel('|s|')
plt.title('График зависимости |s| от n')
plt.show()
```
